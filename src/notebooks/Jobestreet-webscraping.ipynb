```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import pandas as pd
```

```python
jobs = []
search = "developer"
options = Options()
options.add_argument("--headless")
driver = webdriver.Firefox(options=options) #options=options
driver.get('https://www.jobstreet.com.ph/en/job-search/{0}-jobs/'.format(search.replace(" ", "-")))
driver.implicitly_wait(20)
soup_level1 = BeautifulSoup(driver.page_source, 'html.parser')

#finds the number of pages
soup_pages = soup_level1.find_all("option")
pages = int(soup_pages[-1].text)
print(pages)

# loops through all pages
for page in range(1, pages+1):
    driver.get("https://www.jobstreet.com.ph/en/job-search/{0}-jobs/{1}".format(search.replace(" ", "-"),page))
    html = driver.page_source   
    soup = BeautifulSoup(html, 'html.parser')
    job_cards = soup.find_all("div", "z1s6m00 _1hbhsw69y _1hbhsw68u _1hbhsw67e _1hbhsw67q")
    for job in job_cards:
        role = job.find("div", "z1s6m00 l3gun70 l3gun74 l3gun72").text
        company = job.find("a", "z1s6m00 z1s6m0f rqoqz4").text
        loc_sal = job.find_all("span", "z1s6m00 _1hbhsw64y y44q7i0 y44q7i3 y44q7i21 y44q7ih")
        loc = ''
        sal = ''
        if len(loc_sal) > 1:
            loc = loc_sal[0].text
            sal = loc_sal[1].text
        else:
            loc = loc_sal[0].text
        
        print("{0}\n{1}\n{2}\n{3}\n---------------------".format(role, company, loc, sal))
driver.close()
```
